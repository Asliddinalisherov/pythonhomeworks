```python
data_movies = pd.read_csv('movie.csv')
director_name = data_movies.groupby('director_name')['director_facebook_likes'].sum().idxmax()
likes = data_movies.groupby('director_name')['director_facebook_likes'].sum().max()
print(f'{director_name}, {likes} likes')
longest_movies = data_movies.nlargest(5, 'duration')
drs_with_longest_movies = longest_movies['director_name']
print(drs_with_longest_movies)
```
